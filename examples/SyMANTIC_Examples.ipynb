# ***Synthetic Examples***

## ***CaseStudy-1***

```python
from symantic import symantic_model
import numpy as np
import pandas as pd
'''

#############################################################################################################################

Synthetic case study-1

#############################################################################################################################
'''

X = np.random.uniform(1,5,(5,4))
cols = [f'x{i+1}' for i in range(X.shape[1])]
y = 10*X[:,0]/(X[:,1]*(X[:,2]+X[:,3]))

df = pd.DataFrame(X,columns=cols)
df.insert(0,'Target',y)

operators = ['/','+']
rmse,equation,r2 = symantic_model(df,operators=operators,metrics=[0.04,0.995]).fit()

print('\n \n')

print(f'******************************************************** Final Equation:: {equation} ********************************************************')

print('\n \n')
```

```python
'''
If need to see the comparison to the PySR, please uncomment the below code block.
'''

#'''
from pysr import PySRRegressor

model = PySRRegressor(
    niterations=200, #Can be be changed to get good results
    binary_operators=operators,
    unary_operators=[],
    #ncycles_per_iteration=1000,
    maxsize=10, #Can be changed to play around with the complexity of the equations
    elementwise_loss="loss(prediction, target) = sqrt(sum((prediction - target)^2))",

)

model.fit(X, y)

print(model)
#'''
```
